```python
import cv2 #Mengimpor modul cv2 dari OpenCV untuk pemrosesan citra
import numpy as np #Mengimpor modul numpy untuk operasi numerik
from skimage import data #Mengimpor modul data dari skimage untuk citra contoh
import matplotlib.pyplot as plt #Mengimpor modul pyplot dari matplotlib untuk visualisasi
```

```python
#Membaca citra "tank.jpg" dalam mode grayscale
img = cv2.imread("pohon.jpg", cv2.IMREAD_GRAYSCALE) 

#Mendapatkan dimensi citra
row, column = img.shape 

#Membuat array kosong dengan dimensi yang sama dengan citra
img1 = np.zeros((row,column),dtype = 'uint8') 
 
#Menetapkan nilai batas bawah
min_range = 10 
max_range = 60
 
#Looping untuk mengubah piksel menjadi 255 atau 0 berdasarkan batas rentang
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

```python
#Membuat subplot dengan 2 baris dan 2 kolom
fig, axes = plt.subplots(2, 2, figsize=(12, 12)) 
#Mengubah array subplot menjadi 1 dimensi
ax = axes.ravel() 
#menampilkan hasil input citra 
ax[0].imshow(img, cmap=plt.cm.gray)  
ax[0].set_title("Citra Input")  
#Menampilkan histogram citra input
ax[1].hist(img.ravel(), bins=256)  
ax[1].set_title('Histogram Input') 
#menampilkan hasil citra output 
ax[2].imshow(img1, cmap=plt.cm.gray)  
ax[2].set_title("Citra Output")  
#Menampilkan histogram citra output
ax[3].hist(img1.ravel(), bins=256)  
ax[3].set_title('Histogram Output')  
```
