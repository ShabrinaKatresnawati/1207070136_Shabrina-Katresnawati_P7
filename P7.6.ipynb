```python
#import library
import matplotlib.pyplot as plt
#matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
import numpy as np

import cv2
```

```python
#Membaca citra dengan nama file "pohon.jpg" menggunakan fungsi imread.
citra1 = imread(fname="pohon.jpg") 
#Menampilkan dimensi citra
print(citra1.shape)
#Menampilkan citra menggunakan imshow dengan cmap 'gray'.
plt.imshow(citra1, cmap='gray') 
```

```python
kernel = np.array([[-1, 0, -1], 
                   [0, 4, 0], 
                   [-1, 0, -1]])
#Melakukan filtering citra menggunakan kernel dengan filter2D.
citraOutput = cv2.filter2D(citra1, -1, kernel) 
#Membuat objek subplot dengan ukuran 1x2 dan ukuran gambar 12x12.
fig, axes = plt.subplots(1, 2, figsize=(12, 12)) 
#Melakukan ravel pada objek axes untuk mendapatkan array dari semua sumbu subplot.
ax = axes.ravel() 
#Menampilkan citra input  & output pada sumbu indeks.
ax[0].imshow(citra1, cmap='gray') 
ax[0].set_title("Citra Input") 
ax[1].imshow(citraOutput, cmap='gray') 
ax[1].set_title("Citra Output") 
```
